```python
#### Robust Scalar Formula

#      xi - Q2(x)
#  -------------------
#      Q3(X) - Q1(X)

### Inter Quartile Range(IQR)
# Q1 ---> 25%(lower quartile)
# Q2----> 50%(medium quartile)
# Q3 ---->75%(upper quartile)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs=RobustScaler()
rs
```

```python
rs2=rs.fit_transform(data1)
rs2
```

```python
rs2=pd.DataFrame(rs2,columns=data1.columns)
rs2.head()
```

###  Standard Scaler
- most commonly used scalling technique. it scales between 0 to mean value

```python
####  Standard scaler formula
#    xi - mean(x)
#  -------------------
#       std(x)
```

```python
### 1.  import model
### 2.  initialize model into new object
### 3.   fit the model into dataset
### 4.    convert array values into dataframe
```
