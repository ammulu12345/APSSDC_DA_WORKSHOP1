### line plot
- plt.plot()

```python
x_axis=[2,5,8,15]
y_axis=[4,9,16,22]
plt.plot(x_axis,y_axis)
```

```python
x_axis=[2,5,8,15]
y_axis=[4,9,16,22]
plt.plot(x_axis,y_axis,label='line plot')
plt.title("graph of line plotting")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc=7)
```

```python
help(plt.legend)
```

```python
import numpy as np
x=np.linspace(0,20,50)
x
```

```python
plt.plot(x,x**2,color='c',linestyle=':')
plt.plot(x,x**3,linestyle='--')
plt.plot(x,x)
```

```python
help(plt.plot)
```

```python
x2=np.linspace(10,30,50)
x2
```

```python
plt.plot(x2,np.sin(x2),color='green',linestyle='--')
plt.plot(x2,np.cos(x2),color='blue',marker='d')
```
